```python
#GENERATE A RANDOM GAME BOARD

import random
def game_auto_generate() :
    game = []
    
    #random piles number between 2 and 5
    pilesnumber = random.randint(2,5)
    
    #random object number for each piles (between 3 - 8 object per piles)
    for i in range(pilesnumber) :
        game.append(random.randint(3,8))
    
    return game
    
#debug
#print(game_auto_generate())
```

```python
#DISPLAY THE GAME BOARD

def game_display(game) :
    for i in range(len(game)) :
        print("Pile",i,":",game[i])

#Debug
#game = [1,2,3,4,5]
#game_display(game)

#DEBUG W AUTO GENERATE FUNC
game_display(game_auto_generate())
```

```python
#DECLARE WINNING CONDITION

def if_win(game):
    
    #If there's no object left, there's a winner
    if sum(game) == 0 : 
        return True
    #Else, there's no win and the game continue
    else : 
        return False
```

```python
#CALCULATE BOARD'S NIM_SUM USING XOR OPERATOR FOR BINARY (explained in report)

def nimsum_cal(game) :
    sum = 0
    for i in range(len(game)) :
        sum = sum ^ game[i]
    return sum

#Debug
#game = game_auto_generate()
#print(game)
#print(nimsum_cal(game))
```

```python
#IMPLEMENT PLAYER'S MOVE AND COMPUTER MOVE GIVEN THE CONDITION OF GAME BOARD AND NIM-SUM

    #1 : PLAYER MOVE (PLAYER INPUT)
def player_move(game):
    game_display(game)
    print()
    checkval = False
    
    #While loop constantly asking for redo the input when the user have it wrong
    while (checkval == False) :
        pick_pile = input("Choose pile :")
        pick_amount = input("Choose object units to remove :")
        
        #Use try-catch to catch unexpected input that is invalid
        try :
            pick_pile = int(pick_pile)
            pick_amount = int(pick_amount)
        except ValueError :
            print("Input error, please try again")
            checkval = False
            break
        
        #Valid input conditions (within the piles and unit range) 
        if (pick_pile < 0) or (pick_pile > (len(game) - 1)) or (pick_amount < 1) or (pick_amount > game[pick_pile]) :
            print("invalid input, try again")
            checkval = False
        else :
            game[pick_pile] = game[pick_pile] - pick_amount
            checkval = True
    return game

    #2 : COMPUTER PLAY FOLLOWING THE RULES OF MAKING NIM-SUM = 0
def comp_move(game,nim_sum) :
    print()
    game_display(game)
    #If the player finish the turn with nim-sum != 0, try and find the way to remove object in order to make it to 0
    if (nim_sum != 0) :
        for i in range(len(game)):
            #Scanning for possible solution, if the piles value is large enough to remove a "nim-sum" number of unit
            #If yes, take the solution and return the board right away
            if (game[i]^nim_sum < game[i]) :
                #Remove the "nim-sum" number of unit if the piles number is large enought
                print("The computer remove",game[i] - (game[i] ^ nim_sum),"value from pile number",i)
                game[i] = game[i] ^ nim_sum
                return game
            
    #If the player leave nim_sum = 0 to the computer, generate a random move and hope the player make a mistake next time
    else :
        randompile = random.randint(0,len(game)-1)
        game[randompile] -= random.randint(1,game[randompile])
        return game
#debug
#game = game_auto_generate()
#print(game)
#player_move(game)
#print(game)
#comp_move(game,nimsum_cal(game))
 
```

```python
#GAMEPLAY DESIGN
game_continue = True
game = game_auto_generate()

#Loop the game till the winner is found
while (game_continue == True) :
    #Player make a move
    player_move(game)
    #Constant check for winning condition, if player remove the last object, player win, end the game.
    if (if_win(game) == True) :
        print("Player win")
        break
    
    #Computer make a move after player
    comp_move(game,nimsum_cal(game))
     #Constant check for winning condition, If computer remove the last object, computer win, end the game.
    if (if_win(game) == True) :
        print("Computer win")
        break
    
```
